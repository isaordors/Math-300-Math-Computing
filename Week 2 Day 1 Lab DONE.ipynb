The complexity is O(n). This means that it is of order n. For this problem, the code is running for n amount of times. The work is kind of like n + n + n, which is 3n. The power is always 1 so the complexity remains as O(n).

````python
def squares(maximum):
    total = 0
    for i in range(maximum):
        total += i**2
    return total
````

The time complexity is also O(n) for this function since in the worst case, we will run the calculation n amount of times. It is still faster though since we are skiping the containers and going traight to the final answer. This avoids unnessesary calculations.

### Problem 5
Write a recursive function `fibonacci(n)` that computes the $n$-th Fibonacci number. Compute the time complexity of your `fibonacci(n)` function.

````python
def fibonacci(n):
    if n<=1:
        return n
    else:
        return fibonacci(n-2) + fibonacci(n-1)
print(fibonacci(6))
````

yayy! The time complexity is O(2^n) because the calculation will make two recursive calculations (one for n-2 and one for n-1). The bigger the number the more repeated calculations there are, so the growth is exponential.

### Problem 6
Write a recursive function `reverse(string)` that reverses the order of a string. Compute the time complexity of your `reverse()` function

````python
def reverse(string):
    if string == "":
        return ""
    else:
        return reverse(string[1:]) + string[0]

print(reverse("meow"))
````

### Problem 7
Write a recursive function that computes the sum of integers up to input $n$.

````python
def integer(n):
    if n <= 1:
        return 1
    else:
        return n + integer(n-1)
print(integer(6))
````
