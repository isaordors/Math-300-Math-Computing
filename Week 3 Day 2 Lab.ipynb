### Problem 1
Plot the function $f(x)=x^2$ for $x\in [-1,1]$ using `np.linspace()` with 10, 100, 1000 slices. Make sure your plot is as informative as possible.

```python
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
xdata = np.linspace(-1, 1, 10)
ydata = xdata**2
plt.plot(xdata,ydata)
plt.plot(xdata, ydata,label='f(x)')
plt.xlabel('x-values')
plt.ylabel('f(x)')
plt.title('f(x)=$x^2$ graph - 10 slices')
plt.legend()
```

```python
xdata = np.linspace(-1, 1, 100)
ydata = xdata**2
plt.plot(xdata,ydata)
plt.plot(xdata, ydata,label='f(x)')
plt.xlabel('x-values')
plt.ylabel('f(x)')
plt.title('f(x)=$x^2$ graph - 100 slices')
plt.legend()
```

```python
xdata = np.linspace(-1, 1, 1000)
ydata = xdata**2
plt.plot(xdata,ydata)
plt.plot(xdata, ydata,label='f(x)')
plt.xlabel('x-values')
plt.ylabel('f(x)')
plt.title('f(x)=$x^2$ graph - 1000 slices')
plt.legend()
```

```python
x10 = np.linspace(-1, 1, 10)
x100 = np.linspace(-1, 1, 100)
x1000 = np.linspace(-1, 1, 1000)
y10 = x10**2
y100 = x100**2
y1000 = x1000**2
plt.plot(x10, y10, marker='*', label='10 slices')
plt.plot(x100, y100,label='100 slices')
plt.plot(x1000, y1000, label='1000 slices')
plt.legend()

plt.xlabel('x-values')
plt.ylabel('f(x)')
plt.title('f(x)=$x^2$ graph - 10, 100, 1000 slices')
```

### Problem 2
Pick a function $f(x)$ you like and plot some values using `plt.scatter()`. Make sure your plot is as informative as possible.

```python
xdata = np.linspace(-5, 5, 20)
ydata = np.tanh(xdata)
plt.scatter(xdata,ydata,label='f(x)',color='purple',alpha=.40)

plt.xlabel('x-values')
plt.ylabel('f(x)')
plt.title('f(x)=tanh(x) graph - 20 slices')
plt.legend()
```
