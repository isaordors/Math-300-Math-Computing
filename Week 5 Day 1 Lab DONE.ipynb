### Problem 3
Make a 2 by 2 invertible matrix, A.
1. Compute its eigenvectors and plot them using matplotlib
2. Compute A times the eigenvectors and plot them on the same plot.
3. What do you notice?

```python
A = np.array([[1,2],
              [2,6]])
```

```python
eigenvectors=np.linalg.eig(A)[1]
```

```python
v1x,v1y=eigenvectors[0,0],eigenvectors[1,0]
v2x,v2y=eigenvectors[0,1],eigenvectors[1,1]
```

```python
plt.plot([0,v1x],[0,v1y])
plt.plot([0,v2x],[0,v2y])
plt.xlabel("X value")
plt.ylabel("Y value")
plt.title("Eigenvector graph")
plt.show()
```

```python
times=(A@eigenvectors)
print (times)
```

```python
v1x,v1y=eigenvectors[0,0],eigenvectors[1,0]
v2x,v2y=eigenvectors[0,1],eigenvectors[1,1]
v3x,v3y=times[0,0],times[1,0]
v4x,v4y=times[0,1],times[1,1]
```

```python
plt.plot([0, v1x], [0, v1y], color="black", linestyle=":", label="Eigenvector 1")
plt.plot([0, v2x], [0, v2y], label="Eigenvector 2")
plt.plot([0, v3x], [0, v3y], label="A × Eigenvector 1")
plt.plot([0, v4x], [0, v4y], color="purple", linestyle=":", label="A × Eigenvector 2")
plt.xlabel("X value")
plt.ylabel("Y value")
plt.title("Eigenvector and Times Graph")
plt.legend()
plt.show()
```

They all go to the same point (0,0)!!!!!
