## Problem 4
Recall the last problem from yesterday's lab:

"If we list all the natural numbers below 10 that are multiples of 3 or 5 we get 3,5,6, and 9. The sum of these multiples is 23.
Find the sum of all the multiples of 3 or 5 below 1000." 

Now try this problem by defining a function that takes a maximum number as the input and outputs the sum of multiples of 3 or 5 below this number. For example, inputting 10 should output 23.

````python
def function(x):
    y = 0
    for n in range(x):
        if (n % 5 == 0) or (n % 3 == 0):
            y+=n
    return y
function(10)
````

## Problem 6

Define a function

```python
    palindrome(string)
    
```
that takes in a string and outputs whether or not the string is a palindrome. (A palindrome is a word that is the same forward as backward)

````python
def palindrome(string):
    if string == string[::-1]:
        return True
    else:
        return False
palindrome("racecar")
        
````
