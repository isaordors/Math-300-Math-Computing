## Problem 3

A "leap year" is a calendar year where an extra day is added. The rules are as follows:
- If divisible by 400, then leap year
- If divisible by 100, then not a leap year
- If divisible by 4, then leap year
- Otherwise, not a leap year

Write a block of code that takes in the variable `year` and determines if it is a leap year or not.

````python
year = 2073
if year % 400 == 0:
    print ("Leap Year")
elif year % 100 == 0:
    print("Not a Leap Year")
elif year % 4 == 0:
    print ("Leap Year")
else:
    print ("Not a Leap Year")
````

## Problem 5

Write a for-loop that keeps a list of the sum of the first `n` integers as `n` ranges from 1 to 30.

````python
contlist = []
current = 0
for n in range (31):
    current += n
    contlist.append (current)
    print(contlist)
````

## Problem 6
Write a for-loop that counts the number of even integers in a list. You can use the following list as an test:

```python
    
    numbers = [1,13,4,15,2,7,8]
```

````python
count = 0
numbers = [1,13,4,15,2,7,8]
for n in numbers:
    if n % 2 == 0:
        count += 1
print(count)
````
