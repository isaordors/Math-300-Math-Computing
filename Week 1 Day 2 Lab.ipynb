## Problem 3
What does the following print?

```python
print(137 % 12)
```

Explain what the remainder represents.

---

````python
print (137 % 12)
````

The remainder represents how much is left over after dividing as many "groups of 12" as you can out of 137. Since 137 divided by 12 = 11 with a remainder, modulus is essentially telling us that 12 * 11 + 5 is equal to 137, with 5 being the remainder. 

## Problem 4
Determine the values of:

```python
17 / 4
17 // 4
17 % 4
```

How are these quantities related?

---


````python
print (17 / 4)
````

````python
print (17 // 4)
````

````python
print (17 % 4)
````

The first quantity is stating exactly what you get when you divide 17 by 4. It is giving the exact value with no rounding.
The second quantity is essentially rounding the first one. When you divide 17 by 4 originally, you don't get an integer, this computation is providing an integer.
The last quantity is telling us what is left over after we divide 17 by as many groups of 4 as we can. 17 // 4 is 4, and the modulus computation is telling us that 1 is left over without having doing the other computations. It is also telling us that 4 * 4 + 1 is equal to 17.
