```python
contgraph.add_edge('Gabe', 'Alex')
contgraph.add_edge('Gabe', 'Me')
contgraph.add_edge('Me', 'Nicolas')
contgraph.add_edge('Me', 'Marianne')
contgraph.add_edge('Me', 'Miguel')
contgraph.add_edge('Gabe', 'Angel')
contgraph.add_edge('Gabe', 'Dan')
contgraph.add_edge('Dan', 'Angel')
contgraph.add_edge('Miguel', 'Marianne')
contgraph.add_edge('Gabe', 'Nicolas')
contgraph.add_edge('Me', 'Alex')
```

```python
sizes = {
    'Gabe': 3000,
    'Nicolas': 3520,
    'Me': 1500,
    'Alex': 1000,
    'Marianne': 4000,
    'Miguel': 4500,
    'Angel': 1500,
    'Dan': 4500
}

colors = {
    'Gabe': 'brown',
    'Nicolas': 'brown',
    'Me': 'brown',
    'Alex': 'gold',
    'Marianne': 'gold',
    'Miguel': 'brown',
    'Angel': 'gold',
    'Dan': 'brown'
}

edge_sizes = {
    ('Gabe', 'Alex'): 5,
    ('Gabe', 'Me'): 7,
    ('Me', 'Nicolas'): 6,
    ('Me', 'Marianne'): 4,
    ('Me', 'Miguel'): 3,
    ('Gabe', 'Angel'): 5,
    ('Gabe', 'Dan'): 4,
    ('Dan', 'Angel'): 6,
    ('Miguel', 'Marianne'): 6,
    ('Gabe', 'Nicolas'): 3,
    ('Me', 'Alex'): 3
}

plt.title("My Social Network")

nx.draw(
    contgraph,
    with_labels=True,
    node_size=[sizes[node] for node in contgraph.nodes()],
    node_color=[colors[node] for node in contgraph.nodes()],
    width=[edge_sizes.get(edge, edge_sizes.get((edge[1], edge[0]))) for edge in contgraph.edges()]
)
```

### Problem 2

For the dolphin social network:
1. Import the data
2. Draw the network 
3. Use as many `networkx` parameters as possible (that is, in `nx.draw()`, add all the parameters: `node_size`, `node_color`, etc.) such that the network is still presentable and understandable

You can find some information about `nx.draw()` here: https://networkx.org/documentation/stable/reference/generated/networkx.drawing.nx_pylab.draw.html#networkx.drawing.nx_pylab.draw .


```python
dolphincsv = pd.read_csv("DolphinNetwork.csv")
```

```python
dolphincsv
```

```python
edges = [(dolphincsv['Source'].iloc[i], dolphincsv['Target'].iloc[i]) for i in range(len(dolphincsv))]
```

```python
edges
```

```python
dolphin = nx.Graph()
dolphin.add_edges_from(edges)
```

```python
plt.figure(figsize=(12, 8))

nx.draw(
    dolphin,
    with_labels=True,
    node_size=600,
    node_color="lightblue",
    width=1.5,
    font_size=7,
    font_weight="bold",
    alpha=0.75
)

plt.title("Dolphin Social Network")
plt.show()
```
