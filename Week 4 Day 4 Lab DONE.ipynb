```python
Carbons=[1,2,3,4]
Reaction_rate=[5.7*10**8,1.6*10**9,2.3*10**9,4.1*10**9]
```

```python
dataframe = pd.DataFrame({
    'Carbons in primary alcohol(g)': Carbons,
    'Rxn rate with water(g)': Reaction_rate,
})
```

```python
dataframe
```

```python
plt.scatter(Carbons, Reaction_rate)
plt.title('Carbons in primary alcohol(g) vs. rxn rate with water(g)')
plt.xlabel('Number of carbons')
plt.ylabel('Reaction rate')
```

```python
corr = np.corrcoef(Carbons, Reaction_rate)
print(corr)
```

There is a strong positive correlation, with the correlation coefficient being ~0.98. This is because the more carbons there are in a chain, with the first carbon having an attatched hydroxyl group (a primary alcohol), there more energy is required to break the bonds since there are more of them. This means that it will take longer for the 1-butanol (4 carbons) to react than methanol (1 carbon), assuming there is the name molecular amount of each. 

```python
Temperature=[10,15,20,25,30]
Oxygen_Consumption=[4.2,3.5,2.7,1.8,1.2]
```

```python
dataframe = pd.DataFrame({
    'Ambient temperature (C)': Temperature,
    'Oxygen consumption (mL O2/g/h)': Oxygen_Consumption,
})
```

```python
dataframe
```

```python
plt.scatter(Temperature, Oxygen_Consumption)
plt.title('Ambient temperature (C) vs. Oxygen consumption (mL O2/g/h) in warm blooded animals')
plt.xlabel('Temperature (C)')
plt.ylabel('Oxygen consumption (mL O2/g/h)')
```

```python
corr = np.corrcoef(Temperature, Oxygen_Consumption)
print(corr)
```

In this situation, there is a strong negative correlation. The correlation is ~-0.998. This is because warm blooded animals generate their own body heat. When the ambient temperature is colder, warm blooded animals will consume more oxygen since it will help produce more body heat. This allows them to maintain a stable internal temperature despite being in a cold environment.
