### Problem 5: Visualization
1. Choose a numerical column from your dataset and:
    - Plot the corresponding histogram. 
        - What does the distribution look like?

    - Plot the corresponding box plot.
        - Are there any outliers?
        - Explain what you observe.
2. Choose one categorical column from your dataset and:
    - Plot the corresponding bar chart.
        - Which category occurs most frequently?

````python
#1
plt.hist(df["Fare"])
plt.title("Passenger Fares")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
````

This is the histogram for this data. Most paid less than 100. There are a few outliers, with the most notable one being at about 500. The data seems heavily skewed towards the lower end. 

````python
plt.boxplot(df["Fare"])
plt.title("Passenger Fares")
plt.show()
````

There are some outliers. Overall, it seems that the average is quite low, being bellow 100. There are some prices that were above that but nothing too wild. There is one outlier at about 500. There isn't really anyone else around this value so it must have been rare to end up paying that amount of money. 

````python
#2
df["Pclass"].value_counts().plot(kind="bar")
plt.title("Passenger class")
plt.xlabel("Class")
plt.ylabel("Frequency")
plt.show()
````

The 3rd class category occurs most frequently. The 2nd class category occurs least frequently.
