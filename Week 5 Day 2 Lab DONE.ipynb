### Problem 1
Suppose $f(x)=x^2+6x+10$

1. Plot the function to visualize the minimum
2. Find the minimum using `scipy`

````python
x = np.linspace(-50,50,1000)
function = x**2 + 6*x + 10
plt.plot(x,function)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("f(x) graph")
plt.grid()
plt.show()
````

````python
def objective(x):
    return (x**2 + 6*x + 10)
````

````python
result = minimize (objective, x0 = 0)
print(result.x)
````


## Problem 2

Find the maximum of $50x-x^2$.

````python
x = np.linspace(-50,50,1000)
function = 50*x - x**2
plt.plot(x,function)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("f(x) graph")
plt.grid()
plt.show()
````

````python
def f(x):
    return np.sin(x) + x**2
x = np.linspace(-10,10,500)
plt.plot(x,f(x))
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("f(x) graph")
plt.grid()
plt.show()
````

````python
result = minimize (objective, x0 = -10)
print(result.x)
````

````python
result = minimize (objective, x0 = 10)
print(result.x)
````

````python
x = sp.symbols('x')
f=sp.sin(x) + x**2
dv = sp.diff(f, x)
print(sp.diff(f, x))
````

````python
root = sp.nsolve(dv,x,0)
print(root)
````

### Problem 4
Design a function with multiple minima and optimize.

````python
def g(x):
    return np.cos(x)+0.3*x*3
x = np.linspace(-20,20,500)
plt.plot(x,g(x))
plt.grid()
plt.show()
````

````python
result1 = minimize(g,-17)
result2 = minimize(g,0)
result3 = minimize(g,15)

print(result1.x)
print(result2.x)
print(result3.x)
````

````python
x = sp.symbols('x')
g=sp.cos(x) + 0.3*x*3
df = sp.diff(g, x)
print(sp.diff(g, x))
````

````python
root = sp.nsolve(df,x,0)
print(root)
````
