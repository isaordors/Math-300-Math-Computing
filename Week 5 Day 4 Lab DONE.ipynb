```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

# Week 5 Day 4 Lab

### Problem 1
Suppose 

$$\frac{dy}{dt}=1.2y.$$

instead of .5 from lecture. Solve again.

How does the graph change?



```python
def model(t,y):
    return 1.2*y
y0 = [10]
t0=0
t1=10
```

```python
solution = solve_ivp(
    model,
    [t0,t1],
    y0,
    dense_output=True
)
```

```python
t = np.linspace(0,10,200)
y = solution.sol(t)[0]
plt.plot(t,y)
plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Growth")
plt.grid()
plt.show()
```

The graph seems grows at a slower rate initially (though it is kind of hard to tell), but then it seems to grow faster. The population also seems to grow to a much larger number.

## Problem 4

Modify the logistic model by ranging $K$ from 0 to 50 with 100 slices.

How does the solution change?


```python
r = 0.6
K = 100

def logistic(t,y):

    return r*y*(1-y/K)

solution = solve_ivp(
    logistic,
    [0,50],
    [5],
    dense_output=True
)

t = np.linspace(0,50,100)

plt.plot(t,
         solution.sol(t)[0])

plt.axhline(K,
            linestyle="--",
            color="red")

plt.title("Logistic Growth")

plt.show()
```

The solution doesn't seem to change much, though I may just be missing something. 

## Problem 5

Create your own model.

Ideas include

- Predator-prey model
- Cooling of coffee
- Disease spread
- Population growth
- Bank account with continuous interest

Model the system using a differential equation and solve it with `solve_ivp()`

```python
def model(x, y):
    return np.cos(x)*np.exp(-x) + (-1)*np.exp(-x)*np.sin(x)
y0 = [50]
x0 = -50
x1 = 25
solution = solve_ivp(
    model,
    [x0, x1],
    y0,
    dense_output=True
)
```

```python
x = np.linspace(x0,x1,500)
y = solution.sol(x)[0]
plt.plot(x,y)

plt.xlabel("X")
plt.ylabel("Y")

plt.title("Harmonic Oscillator")

plt.grid()

plt.show()
```
